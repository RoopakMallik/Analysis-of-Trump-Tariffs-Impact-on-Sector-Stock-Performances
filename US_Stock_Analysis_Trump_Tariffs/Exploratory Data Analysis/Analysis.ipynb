```python
all_dfs = {
    'Tech': tech_df,
    'Airlines & Logistics': airlines_logistics_df,
    'Auto & Aerospace': auto_aero_df,
    'Goods & Retail': goods_retail_df,
    'Industrial': industrial_df
}

tariff_date_str = '2025-04-02'
tariff_date = pd.to_datetime(tariff_date_str, utc=True)

downward_companies = []

for sector, df in all_dfs.items():
    
    date_col = 'date' if 'date' in df.columns else 'Date'
    df = df.copy()
    
    df['Date'] = pd.to_datetime(df[date_col], utc=True)
    
    post_tariff_data = df[df['Date'] >= tariff_date].copy()
    
    for ticker in post_tariff_data['ticker'].unique():
        ticker_data = post_tariff_data[post_tariff_data['ticker'] == ticker].copy()
        
        if len(ticker_data) < 2:
            continue
        
        price_change = ticker_data['Close'].iloc[-1] - ticker_data['Close'].iloc[0]
        
        volume_change = ticker_data['Volume'].iloc[-1] - ticker_data['Volume'].iloc[0]
        
        if price_change < 0 or volume_change < 0:
            company_name = ticker_data['Company Name'].iloc[0]
            downward_companies.append({
                'Sector': sector,
                'Ticker': ticker,
                'Company Name': company_name,
                'Price Change': price_change,
                'Volume Change': volume_change
            })

downward_companies_df = pd.DataFrame(downward_companies)

downward_companies_df['Volume Change Magnitude'] = downward_companies_df['Volume Change'].abs()

fig = px.scatter(downward_companies_df,
                 x='Price Change',
                 y='Volume Change',
                 size='Volume Change Magnitude',
                 color='Sector',
                 hover_data=['Company Name', 'Ticker'],
                 title='Impact of Tariffs: Price vs. Volume Change',
                 labels={'Price Change': 'Price Change (USD)',
                         'Volume Change': 'Volume Change',
                         'Company Name': 'Company',
                         'Sector': 'Sector'},
                 color_discrete_sequence=px.colors.qualitative.Prism)

fig.update_layout(
    template="plotly_white",
    xaxis_title="Price Change (USD)",
    yaxis_title="Volume Change",
    legend_title="Sector",
    plot_bgcolor='rgba(255,255,255,1)',
    paper_bgcolor='rgba(255,255,255,1)',
    font_color="black"
)

fig.show()
```
