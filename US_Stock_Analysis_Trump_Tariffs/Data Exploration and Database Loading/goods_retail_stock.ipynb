### Loading data to Relational Database

```python
from psycopg2 import connect
pgconn = connect(
    host='localhost',
    database='stock_data_trump_tariffs',
    user='postgres',
    password='4395',
    port='5432'
)
```

```python
pgcursor=pgconn.cursor()
```

```python
from psycopg2.extensions import ISOLATION_LEVEL_AUTOCOMMIT
pgconn.set_isolation_level(ISOLATION_LEVEL_AUTOCOMMIT)
```

```python
pgcursor.execute("CREATE TABLE IF NOT EXISTS goods_retail_stock_data (ticker TEXT, date DATE, open NUMERIC, high NUMERIC, low NUMERIC, close NUMERIC, volume INTEGER);")
```

```python
from psycopg2.extras import execute_values

execute_values(
    pgcursor,
    "INSERT INTO goods_retail_stock_data (ticker, date, open, high, low, close, volume) VALUES %s",
    df[['ticker', 'date', 'Open', 'High', 'Low', 'Close', 'Volume']].values.tolist()
)
```

```python
pgconn.commit()
```
