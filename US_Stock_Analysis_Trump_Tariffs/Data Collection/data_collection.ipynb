```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

tickers = ['F','GM','TSLA','BA','GE','LMT','STLA','NOC']

start_date='2024-01-01'
end_date='2025-05-01'

data={}
for ticker in tickers:
    try:
        stock = yf.Ticker(ticker)
        history = stock.history(start=start_date, end=end_date)
        if not history.empty:
            data[ticker] = history
        else:
            print(f" No data retrieved for {ticker}")
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")

auto_aero_df = pd.concat(data.values(), keys=data.keys(), names=['ticker', 'date'])

auto_aero_df = auto_aero_df.reset_index()

auto_aero_df['Company Name'] = auto_aero_df['ticker'].map({
    'F': 'Ford Motor Company',
    'GM': 'General Motors Company',
    'TSLA': 'Tesla, Inc.',
    'BA': 'Boeing Company',
    'GE': 'General Electric Company',
    'LMT': 'Lockheed Martin Corporation',
    'STLA': 'Stellantis N.V.',
    'NOC': 'Northrop Grumman Corporation'
})
auto_aero_df = auto_aero_df[['ticker', 'date', 'Open', 'High', 'Low', 'Close', 'Volume','Company Name']]

print("Data preview:")
print(auto_aero_df.head())
print(f"Total rows fetched: {len(auto_aero_df)}")

auto_aero_df.to_csv('/Users/roopakmallik/Desktop/Work/Stock_Analysis_Trump_Tariffs/Stock Datasets/auto_aero_stock_data.csv', index=False)
```

```python
import yfinance as yf
import pandas as pd
from datetime import datetime

tickers_info = {
    'DAL': 'Delta Air Lines Inc.',
    'AAL': 'American Airlines Group Inc.',
    'UAL': 'United Airlines Holdings Inc.',
    'LUV': 'Southwest Airlines Co.',
    'ALK': 'Alaska Air Group Inc.',
    'JBLU': 'JetBlue Airways Corp.',
    'UPS': 'United Parcel Service Inc.',
    'FDX': 'FedEx Corp.'
}

start_date = '2024-01-01'
end_date = '2025-05-01'

data = {}

for symbol, name in tickers_info.items():
    try:
        stock = yf.Ticker(symbol)
        history = stock.history(start=start_date, end=end_date)
        if not history.empty:
            history['Company Name'] = name
            data[symbol] = history
        else:
            print(f"No data retrieved for {name} ({symbol})")
    except Exception as e:
        print(f"Error fetching data for {name} ({symbol}): {e}")

airlines_logistics_df = pd.concat(data.values(), keys=data.keys(), names=['ticker', 'Date'])

airlines_logistics_df = airlines_logistics_df.reset_index()
airlines_logistics_df = airlines_logistics_df[['ticker', 'Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Company Name']]

print("Data preview:")
print(airlines_logistics_df.head())
print(f"Total rows fetched: {len(airlines_logistics_df)}")

airlines_logistics_df.to_csv('/Users/roopakmallik/Desktop/Work/Stock_Analysis_Trump_Tariffs/Stock Datasets/airlines_logistics_stock_data.csv', index=False)
```
